```python
all_ingredients = set()
def create_set_ingredients(ingredients_lst):
  global all_ingredients
  #ingredients.add(tuple(ingredients_lst))
  new_set = set(ingredients_lst)
  all_ingredients |= new_set

df['ingredients'].apply(create_set_ingredients)
display(all_ingredients)
```

```python
for ingredient in all_ingredients:
    # Выделяем разряженные признаки на основе элементов словаря ingredients из признака 'ingredients'
    df[ingredient] = df['ingredients'].apply(lambda x: 1 if ingredient in x else 0)
```

```python
def make_list(sr):
  global all_ingredients
  recipe_ingredients = []
  #print(sr)
  for ingredient in all_ingredients:
    #if df[ingredient].iloc[0] == 1:
    if sr[ingredient] == 1:
      recipe_ingredients.append(ingredient)
  return recipe_ingredients



# Создаём пустой список для хранения итоговой структуры
new_recipes = [] 
# Организуем цикл с параметром current_id
for current_id in ids: 
    # Получаем значение соответствующей кухни, применив фильтр по текущему значению параметра цикла к DataFrame;
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0] 
    # Получаем перечень ингредиентов, входящих в состав текущего блюда
    current_ingredients = make_list(df[df['id'] == current_id].iloc[0]) 
    # Создаём текущий словарь
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients} 
    # Добавляем созданный словарь к списку
    new_recipes.append(current_recipe)
```
